```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')


# Let's look at the missing values
print("---Missing Values Before Cleaning---")
print(df.isnull().sum())
```

```python
# Calculate the median of the age column
median_age = df['Age'].median()

# Fill the missing values in age with the median
df['Age'].fillna(median_age, inplace=True)

# Verify that there are no more missing values in age
print("Missing values in 'Age' after imputation:")
print(df['Age'].isnull().sum())
```

```python
# Missing Value Imputation

# Mean Imputation (Best for normally distributed numerical data)
# Example: Using age, assuming it has a normal distribution
if 'Age' in df.columns:
    age_mean = df['Age'].mean()
    df['Age'] = df['Age'].fillna(age_mean)
    print(f"Imputed 'Age' with mean: {age_mean:.2f}")

# Median Imputation (Best for skewed numerical data with outliers)
# Example: Using Fare, which often has extreme outliers
if 'Fare' in df.columns:
    fare_median = df['Fare'].median()
    df['Fare'] = df['Fare'].fillna(fare_median)
    print(f"Imputed 'Fare' with median: {fare_median:.2f}")

# Mode Imputation (Best for catagorical data)
# Example: Using Embarked, filling with the most frequent port
if 'Embarked' in df.columns:
  embarked_mode = df['Embarked'].mode()[0] # [0] grabs the most frequent value
  df['Embarked'] = df['Embarked'].fillna(embarked_mode)
  print(f"Imputed 'Embarked' with mode: {embarked_mode}")

# Verify that no missing values remain in these columns
print("Renaming missing values:")
print(df[['Age', 'Fare', 'Embarked']].isnull().sum())

```

```python
# Impute the 'Age' Column
df['Age'].fillna(df['Age'].mean(), inplace=True)


# Calculate the median of the 'Age' column
age_median = df['Age'].median()

# Replace missing values in the 'Age' column with the median
df['Age'].fillna(age_median, inplace=True)

# Fill the missing values in 'Age' with the calculated median
df['Age'] = df['Age'].fillna(age_median)
# Missing values in 'Age' after imputation
# Print the updated DataFrame
print("Missing values in 'Age' after imputation")
print("---Updated DataFrame---")
print(df)
```

```python
# One-Hot Encoding Catagorical Columns

# Use pd.get_dummies to convert text columns ('Sex', 'Embarked')
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'])

# Display the first few rows to see the new binary columns
print("Data after One-Hot Encoding:")
print(df_encoded.head())
```

```python
encoded_df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
print(encoded_df[['Sex_male']].head())
print("Data after One-Hot Encoding (Drop First):")
print(encoded_df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

encoded_df[['Age', 'Fare']] = scaler.fit_transform(encoded_df[['Age', 'Fare']])

print(encoded_df[['Age', 'Fare']].head())
```

```python
scaler = StandardScaler()

columns_to_scale = ['Age', 'Fare']
encoded_df[columns_to_scale] = scaler.fit_transform(encoded_df[columns_to_scale])
# Display the first few rows to see the scaled data.
print(encoded_df[columns_to_scale].head())
print(encoded_df.head())
```

1. Why is it often better to impute missing values with the median instead of the mean?
 * Robust against outliers: The mean
is heavily dragged up or down by extreme values(like an exceptionally high ticket fare).
 * Handles skewed data: The median marks the exact middle of the data. It gives more realistic "typical" value when data is not perfectly symmetrical.

 * This is a necessity due to most ML algorithms are based from math equations, which cannot read or process raw text string values. This is a way around or straight through that cleans the cell/s.
