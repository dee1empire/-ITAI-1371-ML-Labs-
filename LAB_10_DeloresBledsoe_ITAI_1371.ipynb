```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.8,
                    random_state=42)

# Visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Synthetic Customer Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)

    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('K-Means Clustering Results (k=4)')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns

iris = load_iris()
X_iris = iris.data
y_iris = iris.target

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_iris)

print(f'Explained variance by component: {pca.explained_variance_ratio_}')
print(f'Total variance explained by 2 components: {pca.explained_variance_ratio_}')

pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Species'] = y_iris

plt.figure(figsize=(10, 7))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Species', data=pca_df, palette='viridis', s=100)
plt.title('PCA of Iris Dataset (2 Components)')
plt.show()
```

1. What is the fundamental difference between supervised and unsupervised learning?
***Supervised learning trains on labeled data containing both input features (X) and a corresponding target/correct answer (y) to learn a predictive mapping.
***Unsupervised learning works with unlabeled data containing only features (X) to discover hidden patterns, structures, or natural groupings without a target variable.


2. In the K-Means elbow plot, why don't we just choose the largest possible 'k' to get the lowest inertia?
***Overfitting being a problem. Setting 'k' to the maximum value where 'k'would equal to the total number of data points actually reduces inertia to exactly 0 because every point becomes its own centroid. Diminishing returns another problem, which defeats the entire purpose of clustering, which is to simplify and group data into a smaller number of meaningful, actionable segments. The elbow method indentifies the point where adding more clusters yields diminishing returns.


3. In the PCA results, the first two principal components explained over 95% of the variance. What does this tell you about the original 4-demensional Iris dataset?
***A high-correlation reveals that the original 4-dimensional Iris dataset have significant redundancy and that its features (sepal/petal lengths and widths) are heavily correlated with one another.With dimensionality reduction being 95% of the information and is preserved in 2 components, it proves that a 2d space is good enough to capture, visualize, and model the true structural differences between the plant species without losing critical data.
