```python
# Install required libraries (uncomment if needed)
# !pip install pandas numpy matplotlib scikit-learn

# Import libraries with standard aliases
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Laod the Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

df.head()

# Display dataset information
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

```

```python
# Display the first 5 rows of the dataframe
print('---First 5 Rows---')
print(df.head())

# Display column info and descriptions
print('---Column Descriptions---')
if 'PassengerId' in df.columns:
    print(df.PassengerId.describe())
# Titanic Dataset
print('You are using the Titanic Dataset.')
print("Target Variable: 'Survived' 0 = No, 1 = Yes")
print('Features: Pclass, Sex, Age, SibSp, Parch, Fare, Embarked')
print('description for titanic')

# Display the last 5 rows of the dataframe
print('\n---Last 5 Rows---')
print(df.last)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

plt.figure(figsize=(6, 4))

sns.countplot(data=df, x='Sex', hue='Survived')

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Survival Rate by Passenger Class
print('---Survival Rate by Passenger Class---')
plt.figure(figsize=(10, 6))

sns.barplot(data=df, x='Pclass', y='Survived', errorbar=None, palette='muted')

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class (1st, 2nd, 3rd)")
plt.ylabel("Survival Rate (Percentage)")

plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Titanic Survival Rate by Age vs Passenger Class
print('---Titanic Survival Rate by Age vs Passenger Class---')

# Create the FacetGrid

g = sns.FacetGrid(df, col='Pclass', row="Sex",hue='Survived', height=4, aspect=1.2)

# Add count plots to each facet
g.map_dataframe(sns.countplot, x="Embarked", order=["C", "Q", "S"])


# Add legend and titles
g.add_legend(title='Survived')
g.set_axis_labels("Port of Embarkation", "Count")
g.set_titles(row_template="{row_name} Class", col_template="{col_name} Sex")

plt.subplots_adjust(top=0.9)
g.fig.suptitle('Titanic Survival Rate by Age vs Passenger Class')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# Titanic Survival relationships across multiple categories
g.map(sns.histplot, "Age", bins=20, color="steelblue")

# Add Titles and labels
g.set_axis_labels("Age", "Number of Passengers")
g.set_titles(row_template="Gender: {row_name}", col_template="Survived: {col_name}")

plt.subplots_adjust(top=0.88)
g.fig.suptitle("Age Distribution by Gender and Survival Status")

plt.show()
```

I created two visualizations to explore the Titanic dataset. The first visualization was a countplot comparing passengers, seperated by wether they survived or not.

The first visualization shows that female passengers survival rate was much higher than that of male passengers. The second visualization showed that survival patterns differ throughout both genders. While female passengers showed a high survival rate, the males had more deaths reguardless of age. The FacetGrid showed that the effect on all age groups, although the younger age groups had a better survival rate outcome.

My Hypothesis is that gender, passenger class, and age will be important features for predicting passenger survival.
